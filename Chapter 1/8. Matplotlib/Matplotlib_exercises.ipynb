### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here

fig, ax = plt.subplots(figsize=(8, 10), subplot_kw=dict(aspect="equal"))
Total_facecreams = data["facecream"].sum()
Total_facewash = data["facewash"].sum()
Total_toothpaste = data["toothpaste"].sum()
Total_bathingsoaps = data["bathingsoap"].sum()
Total_shampoo  = data["shampoo"].sum()
Total_moisturizer = data["moisturizer"].sum()
Total_units = data["total_units"].sum()
my_labels = data.columns[1:-2].values.tolist()

Total_units_sold = [Total_facecreams,Total_facewash,Total_toothpaste,Total_bathingsoaps,Total_shampoo,Total_moisturizer]

def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} units )".format(pct, absolute)


wedges, texts, autotexts = ax.pie(Total_units_sold, autopct=lambda pct: func(pct, Total_units_sold),textprops=dict(color="w"))

ax.legend(wedges, my_labels,
          title="Products list",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Matplotlib bakery: A pie")

plt.show()
```
